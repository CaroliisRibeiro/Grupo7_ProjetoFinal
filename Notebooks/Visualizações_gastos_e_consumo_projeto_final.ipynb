### Passo 1: Configuração do Ambiente

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Visualizando % de visitantes que pernoitam

```python
# Dados
perc_pernoitam = 95.86
total_visitantes = 169

# Cálculos
qtd_pernoitam = round(total_visitantes * (perc_pernoitam / 100))
qtd_nao_pernoitam = total_visitantes - qtd_pernoitam

dados_pernoite = {
    "Categoria": ["Pernoitam", "Não pernoitam"],
    "Valor": [perc_pernoitam, 100 - perc_pernoitam],
    "Quantidade": [qtd_pernoitam, qtd_nao_pernoitam]
}

# Cores
cores = ["#2c3e50", "#a0a0a0"]

# Gráfico de pizza
fig, ax = plt.subplots(figsize=(7,7), facecolor="#f8f9fa")
wedges, texts, autotexts = ax.pie(
    dados_pernoite["Valor"],
    labels=[f"{cat}\n({qtd} visitantes)" for cat, qtd in zip(dados_pernoite["Categoria"], dados_pernoite["Quantidade"])],
    autopct='%1.1f%%',
    colors=cores,
    startangle=140,
    textprops={'fontsize': 11, 'color': '#2c3e50'}  # só mudamos a cor da letra aqui
)

# Estilo do título
plt.title("% de Visitantes que Pernoitaram", fontsize=14, fontweight="bold", color="#2c3e50", pad=20)

# Configurações do gráfico
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### Médias de gastos dos visitantes

```python
#Dados
dados_gastos = {
    "Gasto": [
        "Hospedagem", "Atrativos", "Alimentação", "Transporte",
        "Fantasia", "Compras", "Ingressos", "Outros"
    ],
    "Média": [
        346.96,   # Hospedagem
        358.21,   # Atrativos
        540.71,   # Alimentação
        169.12,   # Transporte
        69.51,    # Fantasia
        129.57,   # Compras
        71.20,    # Ingressos
        14.09     # Outros
    ]
}

df_gastos = pd.DataFrame(dados_gastos)

# Exibir tabela
print("Médias de Gastos dos Visitantes:")
display(df_gastos)

# Gráfico de barras

cores = ["#2c3e50", "#a0a0a0"]  # azul escuro e cinza médio
# Alternando cores para cada barra
barra_cores = [cores[i % len(cores)] for i in range(len(df_gastos))]

plt.figure(figsize=(10,6))
plt.bar(df_gastos["Gasto"], df_gastos["Média"], color=barra_cores)
plt.title("Médias de Gastos dos Visitantes", fontsize=14, fontweight="bold", color="#2c3e50")
plt.xlabel("Categoria", fontsize=12, color="#2c3e50")
plt.ylabel("Valor Médio (R$)", fontsize=12, color="#2c3e50")
plt.xticks(rotation=45, fontsize=10, color="#2c3e50")
plt.yticks(fontsize=10, color="#2c3e50")
plt.grid(axis='y', linestyle='--', alpha=0.7)  # linhas de grade para facilitar leitura
plt.tight_layout()
plt.show()
```

### **Análise de resultados e insights estratégicos**


🎭 1. Perfil e Comportamento dos Visitantes

- Alta taxa de pernoite: média de pernoite >4 → oportunidade de fomentar pacotes turísticos (hotéis + experiências culturais).

- Meios de hospedagem: predominância de casa de amigos/parentes → sinal de que a rede hoteleira não está sendo totalmente aproveitada. Ação: incentivar hotéis e Airbnbs com promoções ou parcerias oficiais.

🍹 2. Gastos dos Visitantes

Categorias que mais puxam os gastos: Alimentação, Atrativos e Hospedagem.

 - A Alimentação é o principal gasto, indicando a importância crítica da infraestrutura de food service. Atrativos e Hospedagem em seguida, confirmando que o público investe em experiências, lazer e estadia na cidade.

--> Ação: Reforçar a fiscalização e o licenciamento de ambulantes e restaurantes e promover pacotes turísticos (Atrativos + Hospedagem) para consolidar a receita.

- Transporte e Compras mantêm peso intermediário.

--> Ação: Ampliar as rotas de transporte, fortalecer parcerias com apps de corridas e ampliar as rotas de ônibus durante o evento, além de incentivar feiras de artesanato e produtos temáticos em locais de fácil acesso.

- Ingressos e Fantasias com os menores valores entre as despesas relevantes.

  Contexto: Ingressos e Fantasias têm gastos menores, indicando que o foco pode estar mais na folia de rua gratuita e menos em eventos fechados de alto custo ou na compra de fantasias caras.

--> Ação: Avaliar a precificação de eventos pagos e/ou promover eventos de acesso mais fácil para o público.


🚎 3. Transporte e Mobilidade

- Forte uso de aplicativos (Uber/99) → mostra conveniência, mas pode gerar trânsito e encarecimento em horários de pico.

--> Ação: prefeitura poderia reforçar transporte coletivo com rotas exclusivas para blocos e pontos de embarque organizados, linhas de ônibus 24h, pontos de apoio para táxis/Uber, incentivo ao uso de bikes e patinetes.
