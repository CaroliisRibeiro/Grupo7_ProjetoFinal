##  Frequência das faixas


```python
import numpy as np
import pandas as pd

# colunas numéricas de gasto
gastos = [
    "gastohospedagem",
    "gastoatrativos",
    "gastoalimentacao",
    "gastotransporte",
    "gastofatansias",
    "gastocompras",
    "gastoingressos",
    "gastooutros",
    "gastototal",
    "gastopercapto",
    "gastomediodiario",
    "gastomediodiariopercapto",
]

# colunas de FAIXA
faixas = [
    "faixagastohospedagem",
    "faixagastoatrativos",
    "faixagastoalimentacao",
    "faixagastotransporte",
    "faixagastofatansias",
    "faixagastocompras",
    "faixagastoingressos",
    "faixagastooutros",
    "faixagastototal",
    "faixagastopercapto",
    "faixagastomediodiario",
    "faixagastomediodiariopercapto",
]

# Função para converter "R$ 1.234,56" -> 1234.56 e remover lixo
def to_number(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    s = str(x).strip()
    # remove tudo que não for dígito, vírgula, ponto, sinal
    s = pd.Series(s).str.replace(r"[^0-9,.\-]", "", regex=True).iloc[0]
    # padrão BR: troca milhar '.' por nada e decimal ',' por '.'
    s = s.replace(".", "").replace(",", ".")
    return pd.to_numeric(s, errors="coerce")

# 4) Converte TODAS as colunas de gasto para numérico
for g in gastos:
    if g in df.columns:
        df[g] = df[g].apply(to_number)

# 5) Define bins e rótulos
bins = [-0.01, 0, 100, 500, 1000, float("inf")]
labels = ["Muito baixo (zero)", "Baixo", "Médio", "Alto", "Muito alto"]

# 6) Preenche as colunas de faixa
for g, f in zip(gastos, faixas):
    if g in df.columns and f in df.columns:
        df[f] = pd.cut(df[g], bins=bins, labels=labels, right=True)

# 7) Checagens rápidas (opcional)
# Veja se sobrou alguma linha que não virou número em cada coluna de gasto
na_report = {g: int(df[g].isna().sum()) for g in gastos if g in df.columns}
print("Valores não numéricos (após conversão) por coluna de gasto:", na_report)

# Frequência das faixas para uma coluna exemplo
if "faixagastototal" in df.columns:
    print(df["faixagastototal"].value_counts(dropna=False))
```

#Coluna Renda e faixarenda Com Rótulos

```python
# Dicionário de mapeamento
map_renda = {
    "Ate 1 SM (R$ 1.30200)" : "Até 1 SM",
    "Entre 1 SM e 3 SM (R$ 1.30201 a R$3.90600)" : "Entre 1 e 3 SM",
    "Entre 3 SM e 5 SM (R$3.906 01 a R$6.51000)" : "Entre 3 e 5 SM",
    "Entre 5 SM e 7 SM (R$6.51001 a R$ 9.11400)" : "Entre 5 e 7 SM",
    "Entre 7 SM e 10 SM (R$9.11401 a R$13.02000)" : "Entre 7 e 10 SM",
    "Acima de 10 SM (R$ 13.02001)" : "Acima de 10 SM"
}

# Preenche a coluna existente 'faixarenda'
df["faixarenda"] = df["renda"].map(map_renda)

# Confere resultado
print(df[["renda", "faixarenda"]].head(15))
```

## Criar valores numéricos aproximados para cada faixa


```python
map_valor = {
    "Até 1 SM" : 1302,                           # valor aproximado do teto
    "Entre 1 e 3 SM" : (1302 + 3906) / 2,        # média da faixa
    "Entre 3 e 5 SM" : (3906 + 6510) / 2,
    "Entre 5 e 7 SM" : (6510 + 9114) / 2,
    "Entre 7 e 10 SM" : (9114 + 13020) / 2,
    "Acima de 10 SM" : 13020                     # pode ser o mínimo ou extrapolar
}

# Preenche a coluna existente 'rendamedia' com os valores numéricos
df["rendamedia"] = df["faixarenda"].map(map_valor)

# Confere resultado
print(df[["faixarenda", "rendamedia"]].head(15))
```
